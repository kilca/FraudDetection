```python
import numpy as np
from scipy.sparse import csr_matrix,csc_matrix, lil_matrix
from scipy.io import savemat, loadmat
import pandas as pd
```

# Amazon_mp3

```python
def changeCond(line,dict):
    ind = line.find(":")
    if (ind != -1):
        #print("we set : ",line[1:ind-1], " to ",line[ind+1:])
        dict[line[1:ind-1]] = line[ind+1:]

#df = pd.DataFrame(columns=['A'])
df = pd.DataFrame()

d = {}
with open('data_in/amazon_mp3','r') as file:
    for line in file:
        line = line.strip()
        if (line == ""):
            df = df.append(d, ignore_index=True)
        if(line.startswith("#####")):
            d = {}
            d['num'] = line.split("#####")[1]
        else:
            changeCond(line,d)
        
```

##### Exportation en CSV

```python
df.to_csv('data_out/amazon_mp3.csv', index=False)
```
